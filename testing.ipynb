```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
import os
os.environ["CUDA_VISIBLE_DEVICES"]="0"
os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"
gpu_options = tf.GPUOptions(per_process_gpu_memory_fraction=0.50)
```

```python
xtest = np.load('xtest.npy')
ytest = np.load('ytest.npy')
```

```python
_,height,width = xtest.shape
channels = 1
classes=2
xtest = np.reshape(xtest,newshape=(-1,height,width,channels))
with tf.Session(config=tf.ConfigProto(gpu_options=gpu_options)) as sess:
    ytest = sess.run(tf.one_hot(ytest,depth=classes))
```

```python
print(xtest.shape)
print(ytest.shape)
```

```python
with tf.Session() as sess:
    saver = tf.train.import_meta_graph("saved_models/catvdog.ckpt.meta")
    saver.restore(sess,tf.train.latest_checkpoint("saved_models/"))
    graph = tf.get_default_graph()    
    y_pred = graph.get_tensor_by_name("y_pred:0")
    x_true = graph.get_tensor_by_name("x_true:0")
    p = sess.run(y_pred,feed_dict={x_true:xtest})
    prediction = sess.run(tf.argmax(p,axis=1))
```

```python
for i in range(len(prediction)):
    plt.figure()
    plt.imshow(np.squeeze(xtest[i]))
    plt.title(prediction[i])
```
